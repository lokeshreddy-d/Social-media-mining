```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the original CSV file
df = pd.read_csv('/content/final_toxic_dataset.csv')
#print(f"length of dataset: {len(df)}")

toxicity_counts = []
categories = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

# Count the number of 1s in each specified column and store in a dictionary
toxicity_counts = {col: (df[col] == 1).sum() for col in categories}

# non_toxicity_counts = (df[categories] == 0).all(axis=1).sum()
# toxicity_counts["non_toxic"] = non_toxicity_counts

print(toxicity_counts)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(toxicity_counts.keys(), toxicity_counts.values(), color='red')

plt.xticks(rotation=45)
plt.ylabel('Counts')
plt.title('Counts of Toxic & Nontoxic Tweets')
plt.savefig('/content/tweet_counts_plot.jpg', dpi=300, format='jpg')
plt.show()
```
